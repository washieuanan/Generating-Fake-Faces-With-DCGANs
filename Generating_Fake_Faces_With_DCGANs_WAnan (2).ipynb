# The Production of Fake Faces with Deep Convolutional Generative Adversarial Networks (DCGANs)

Generative adversarial networks (also known as GANs) are algorithmic architectures that pit two neural networks against each other in order to generate new, synthetic instances of data, that (hopefully) can pass for real data. Generative Adversarial Networks are commonly used in image generation, video generation, and voice generation. 

GANs are made up of two components: 

1. Generator: Generates new data instances
2. Discriminator: Tries to distinguish the generated data from the fake data in the real dataset.

While training, both the generator and discriminator begin together. Through training, the generator learns to shape the random distribution through the training epochs: 



---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
In the paper that had introduced GANs (link in reference), the generator tries to minimize the following function while the discriminator tries to maximize it: 



In this loss function: 

D(x) is the discriminator's estimate of the probability that real data instance (x) is real. 

Ex is the expected value over all of the data instances. 

G(z) is the generator's output when given noise z. 

D(G(z)) is the discriminator's estimate of the probability that a fake instance is real. 

Ez is the expected value over all random inputs that are going into the generator (the expected value over all generated fake instances G(z)). 

The generator cannot directly affect log(D(x)) term in the function. This means that the generator minimizes the loss, thereby making it equivalent to minimizing log(1-D(G(z))).

# Architecture of the Model

The core of a DCGAN architecture utilizes a standard CNN architecture on the discriminative model. For the generator, convolutions are replaced with upconvolutions. This allows for the representation at each layer of the generator to become successively larger. It helps map the images from a low-dimensional latent vector onto a high-dimension image. 

Both the generator and the discriminator will utilize batch normalization. 



```python
noise_shape = 100
```

```python
#Generator
generator=Sequential()
generator.add(Dense(4*4*512,input_shape=[noise_shape]))
generator.add(Reshape([4,4,512]))
generator.add(Conv2DTranspose(256, kernel_size=4, strides=2, padding="same"))
generator.add(LeakyReLU(alpha=0.2))
generator.add(BatchNormalization())
generator.add(Conv2DTranspose(128, kernel_size=4, strides=2, padding="same"))
generator.add(LeakyReLU(alpha=0.2))
generator.add(BatchNormalization())
generator.add(Conv2DTranspose(64, kernel_size=4, strides=2, padding="same"))
generator.add(LeakyReLU(alpha=0.2))
generator.add(BatchNormalization())
generator.add(Conv2DTranspose(3, kernel_size=4, strides=2, padding="same",
                                 activation='sigmoid'))
```

```python
generator.summary()
```

```python
#Discriminator
discriminator=Sequential()
discriminator.add(Conv2D(32, kernel_size=4, strides=2, padding="same",input_shape=[64,64, 3]))
discriminator.add(Conv2D(64, kernel_size=4, strides=2, padding="same"))
discriminator.add(LeakyReLU(0.2))
discriminator.add(BatchNormalization())
discriminator.add(Conv2D(128, kernel_size=4, strides=2, padding="same"))
discriminator.add(LeakyReLU(0.2))
discriminator.add(BatchNormalization())
discriminator.add(Conv2D(256, kernel_size=4, strides=2, padding="same"))
discriminator.add(LeakyReLU(0.2))
discriminator.add(Flatten())
discriminator.add(Dropout(0.5))
discriminator.add(Dense(1,activation='sigmoid'))
```

```python
discriminator.summary()
```

```python
#DCGAN (Combined Model)
GAN = Sequential([generator,discriminator])
```

```python
discriminator.compile(optimizer='adam',loss='binary_crossentropy')
discriminator.trainable = False
```

```python
GAN.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
GAN.summary()
```

# **Training**

```python
epochs = 300  
batch_size = 128
```

```python
D_loss=[] #list to collect loss for the discriminator model
G_loss=[] #list to collect loss for generator model
```

```python
with tf.device('/gpu:0'):
    for epoch in range(epochs):
        print(f"Currently on Epoch {epoch+1}")
    
    # For every batch in the dataset
        for i in range(X_train.shape[0]//batch_size):
        
            if (i)%100 == 0:
                print(f"\tCurrently on batch number {i} of {len(X_train)//batch_size}")
            
            noise=np.random.uniform(-1,1,size=[batch_size,noise_shape])
        
            gen_image = generator.predict_on_batch(noise)
        
            train_dataset = X_train[i*batch_size:(i+1)*batch_size]
            #train on real image
            train_label=np.ones(shape=(batch_size,1))
            discriminator.trainable = True
            d_loss1 = discriminator.train_on_batch(train_dataset,train_label)
        
        #train on fake image
            train_label=np.zeros(shape=(batch_size,1))
            d_loss2 = discriminator.train_on_batch(gen_image,train_label)
        
        
            noise=np.random.uniform(-1,1,size=[batch_size,noise_shape])
            train_label=np.ones(shape=(batch_size,1))
            discriminator.trainable = False
        #train the generator
            g_loss = GAN.train_on_batch(noise, train_label)
            D_loss.append(d_loss1+d_loss2)
            G_loss.append(g_loss)
        
         
        if epoch % 5 == 0:
            samples = 10
            x_fake = generator.predict(np.random.normal(loc=0, scale=1, size=(samples,100)))

            for k in range(samples):
                plt.subplot(2, 5, k+1)
                plt.imshow(x_fake[k].reshape(64,64,3))
                plt.xticks([])
                plt.yticks([])

        
            plt.tight_layout()
            plt.show()
        print('Epoch: %d,  Loss: D_real = %.3f, D_fake = %.3f,  G = %.3f' %   (epoch+1, d_loss1, d_loss2, g_loss))        
print('Training is complete')
```
